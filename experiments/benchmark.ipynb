```python
import numpy as np
import pickle
import os
```

```python
def get_all_files(directory):
    files = []
    for root, dirs, filenames in os.walk(directory):
        for filename in filenames:
            if filename == "scores.pkl":
                files.append(os.path.join(root, filename))
    return files
```

```python
best = []

with open("best_solutions.txt", "r") as f:
    for line in f:
        best.append(int(line.strip()))
best
```

```python
directory = "../out/combined_visit/benchmarks/chao"
# directory = "../out/benchmarks/chao"
files = get_all_files(directory)
files.sort()
for file in files:
    print(file)
```

```python
counter = 0
counters = []
avgs = []
avg = []
avg2 = []
i = 1

for file, b in zip(files, best):
    scores = []
    with open(file, "rb") as f:
        while True:
            try:
                scores.append(pickle.load(f))
            except EOFError:
                break
    
    score = max([s[0] for s in scores])
    if f"p{i}" not in file:
        print(f"p{i}:", np.mean(avg), counter, np.mean(avg2))
        i += 1
        avgs.append(np.mean(avg))
        avg = []
        counters.append(counter)
        counter = 0

    # print(file, b, score)
    # print(b, score)
    if score < b:
        avg.append((b - score) / b)
    else:
        counter += 1
    avg2.append(np.mean(np.array([score[1] for score in scores])))

print(f"p{i}:", np.mean(avg), counter, np.mean(avg2))
avgs.append(np.mean(avg))
counters.append(counter)
print("Average:", np.mean(avgs))
print("Counters:", np.sum(counters) + 13)
print("Num files:", len(files) + 13)
```
