```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Get results for budget 200

num_agents = ['2', '3', '6', '8']
budget = '200'
mean200 = {}
std200 = {}
num_iters200 = {}

for n in num_agents:
    with open(f'../tests/it_time_{n}_bdg_{budget}.txt', 'r') as f:
        lines = f.readlines()
        runtimes = [float(line.split()[-1]) for line in lines]
        mean200[n] = np.mean(runtimes)
        std200[n] = np.std(runtimes)
        num_iters200[n] = len(runtimes)

mean200, std200, num_iters200
```

```python
# Get results for budget 700

mean700 = {}
std700 = {}
num_iters700 = {}

for n in num_agents:
    with open(f'../tests/it_time_{n}.txt', 'r') as f:
        lines = f.readlines()
        runtimes = [float(line.split()[-1]) for line in lines]
        mean700[n] = np.mean(runtimes)
        std700[n] = np.std(runtimes)
        num_iters700[n] = len(runtimes)

mean700, std700, num_iters700
```

```python
fig, ax = plt.subplots()

# Plotting error bars with candle-like appearance
ax.errorbar([int(n) for n in num_agents], [mean200[n] for n in num_agents], yerr=[std200[n] for n in num_agents], fmt='o', color='blue', label='Budget 200', capsize=5, capthick=2, elinewidth=2)
ax.errorbar([int(n) for n in num_agents], [mean700[n] for n in num_agents], yerr=[std700[n] for n in num_agents], fmt='o', color='red', label='Budget 700', capsize=5, capthick=2, elinewidth=2)

# Setting x-axis ticks to only show numbers that have data
ax.set_xticks([int(n) for n in num_agents])

ax.set_xlabel('Number of agents')
ax.set_ylabel('Runtime (s)')
ax.set_title('Runtime of an iteration for different number of agents')
ax.legend()

plt.grid(True)

plt.show()
```
